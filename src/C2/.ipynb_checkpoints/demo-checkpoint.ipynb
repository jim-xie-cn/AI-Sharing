```python
import sys
#!{sys.executable} -m pip install seaborn==0.9.0
import seaborn
import random
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import adfuller
from statsmodels.stats.diagnostic import acorr_ljungbox
from statsmodels.tsa.arima_model import ARIMA
from scipy.stats import chi2_contingency
from statsmodels.sandbox.stats.multicomp import multipletests
from itertools import combinations
#sns.set(rc={"figure.figsize": (24, 12)})
import json
plt.style.use({'figure.figsize':(24, 12)})
sns.set_style('dark')
sns.set_context("poster")
pd.set_option('display.float_format',lambda x : '%.2f' % x)
plt.rcParams['axes.unicode_minus'] = False
```

# Prepare train data

```python
df = pd.read_csv('./US/us_states_covid19_daily.csv')
del df['hash']
del df['commercialScore']
del df['negativeRegularScore']
del df['negativeScore']
del df['positiveScore']
del df['score']
del df['grade']
del df['hospitalizedIncrease']
del df['deathIncrease']
del df['hospitalizedCumulative']
del df['hospitalizedCurrently']
del df['inIcuCurrently']
del df['inIcuCumulative']
del df['onVentilatorCurrently']
del df['onVentilatorCumulative']
del df['recovered']
del df['lastUpdateEt']
del df['dateModified']
del df['checkTimeEt']
del df['dateChecked']
df['date'] = df['date'].astype(str)
df.fillna(value=0,inplace=True)
df.head(5)
```

```python
def GetRawSamples():
    df2 = pd.DataFrame()
    df2['date'] = df['date']
    df2['state'] = df['state']
    df2['positive']=df['positive']
    df2['positiveIncrease']=df['positiveIncrease']
    df2['test'] = df['total']
    df2['testIncrease'] = df['totalTestResultsIncrease']
    df2.sort_values('date',ascending=True,inplace=True)
    df2 = df2[df2['date']>='20200401']
    df2 = df2.reset_index()
    del df2['index']
    return df2

def GetSamples(state = None):
    dr = GetRawSamples()
    if state:
        dr = dr[dr['state']==state]
    else:
        dr = dr.groupby(['date'])['positive','positiveIncrease','test','testIncrease'].agg('sum')
        dr['state'] = 'ALL'
    dr.sort_values('date',ascending=True,inplace=True)
    dr = dr.reset_index()
    return dr
```

```python
df0 = GetSamples()
train_data = df0['positiveIncrease']
train_date = df0['date']
```

```python
plt.style.use({'figure.figsize':(32, 8)})
plot_acf(train_data).show()
```

```python
plt.style.use({'figure.figsize':(32, 8)})
plot_pacf(train_data).show()
```

# Train

```python
model_ar = ARIMA(train_data, order=(3, 0, 0))
AR_model = model_ar.fit()
plt.plot(train_data)
plt.plot(AR_model.fittedvalues, color='red')
rmse = np.sqrt(sum((AR_model.fittedvalues-train_data)**2)/train_data.size)
plt.title('ARIMA RMSE:%.4f' % rmse)
plt.show()
```

```python
model_arma = ARIMA(train_data, order=(3, 0, 2))
ARMA_model = model_arma.fit(disp=-1)
plt.plot(train_data)
plt.plot(ARMA_model.fittedvalues, color='red')
rmse = np.sqrt(sum((ARMA_model.fittedvalues-train_data)**2)/train_data.size)
plt.title('ARIMA RMSE:%.4f' % rmse)
plt.show()
```
