# Time Series Forecasting

# Prepare train data

```python
df = pd.read_csv('./US/us_states_covid19_daily.csv')
del df['hash']
del df['commercialScore']
del df['negativeRegularScore']
del df['negativeScore']
del df['positiveScore']
del df['score']
del df['grade']
del df['hospitalizedIncrease']
del df['deathIncrease']
del df['hospitalizedCumulative']
del df['hospitalizedCurrently']
del df['inIcuCurrently']
del df['inIcuCumulative']
del df['onVentilatorCurrently']
del df['onVentilatorCumulative']
del df['recovered']
del df['lastUpdateEt']
del df['dateModified']
del df['checkTimeEt']
del df['dateChecked']
df['date'] = df['date'].astype(str)
df.fillna(value=0,inplace=True)
df.head(5)
```

```python
def GetRawSamples():
    df2 = pd.DataFrame()
    df2['date'] = df['date']
    df2['state'] = df['state']
    df2['positive']=df['positive']
    df2['positiveIncrease']=df['positiveIncrease']
    df2['test'] = df['total']
    df2['testIncrease'] = df['totalTestResultsIncrease']
    df2.sort_values('date',ascending=True,inplace=True)
    df2 = df2[df2['date']>='20200401']
    df2 = df2.reset_index()
    del df2['index']
    return df2

def GetSamples(state = None):
    dr = GetRawSamples()
    if state:
        dr = dr[dr['state']==state]
    else:
        dr = dr.groupby(['date'])['positive','positiveIncrease','test','testIncrease'].agg('sum')
        #dr = dr.to_frame().reset_index()
        dr['state'] = 'ALL'
    dr.sort_values('date',ascending=True,inplace=True)
    dr = dr.reset_index()
    return dr
```

```python
df0 = GetSamples()
df1 = GetSamples('NY')
df2 = GetSamples('CA')
df3 = GetSamples('IL')
#sns.lineplot(x='date', y='positiveIncrease', hue="state", markers=True, data=df1)
train_data = df0['positiveIncrease']
train_date = df0['date']
```

# 平稳性检测 

```python
plt.style.use({'figure.figsize':(32, 8)})
plot_acf(train_data).show()
```

```python
plt.style.use({'figure.figsize':(32, 8)})
plot_pacf(train_data).show()
```

```python
print('stable test ADF : ')
def unit_root_test(data):
    result = adfuller(data,maxlag=1,regression='ct')
    acf_value = result[0]
    p_value = "%0.20f"%result[1]
    level_value = result[4]
    print("acf:",acf_value)
    print("p:",p_value)
    print("level:",json.dumps(level_value,indent=4))

unit_root_test(train_data)
```

```python
print("noise : ")
result  = acorr_ljungbox(train_data, lags=1)
noise_value = "%0.20f"%result[1]
print(noise_value)
#for item in result[1]:
#     noise_value = "%0.20f"%item
#     print(noise_value)
```

# 模型选择

```python
model_ar = ARIMA(train_data, order=(3, 0, 0))
AR_model = model_ar.fit()
plt.plot(train_data)
plt.plot(AR_model.fittedvalues, color='red')
rmse = np.sqrt(sum((AR_model.fittedvalues-train_data)**2)/train_data.size)
plt.title('ARIMA RMSE:%.4f' % rmse)
plt.show()
```

```python
model_arma = ARIMA(train_data, order=(3, 0, 2))
ARMA_model = model_arma.fit(disp=-1)
plt.plot(train_data)
plt.plot(ARMA_model.fittedvalues, color='red')
rmse = np.sqrt(sum((ARMA_model.fittedvalues-train_data)**2)/train_data.size)
plt.title('ARIMA RMSE:%.4f' % rmse)
plt.show()
```

# prediction

```python
start = train_data.size 
end = start 
start_date = train_date.iloc[start - 1]
predict_data=AR_model.predict(start,end)             
#forecast=AR_model.forecast(5)                              
# 绘制原数据和预测数据对比图
print("开始日期：",start_date)
for i in range(0,predict_data.size):
    print(i+1,"天后新增：",int(predict_data.iloc[i]))
```
