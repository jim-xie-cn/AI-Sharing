# 人口分布

```python
df = pd.read_csv('./COVID19/population_india_census2011.csv')
```

```python
df['Jim_Area'] = df['Area'].str.partition('km')[0].str.replace(',','').astype(int)
df['Jim_Density'] = df['Density'].str.partition('/km')[0].str.replace(',','').astype(float)
df['Population'] = df['Population'].astype(int)
df.head(2)
```

```python
plt.style.use({'figure.figsize':(32, 16)})
sns.barplot(y="State / Union Territory", x="Population", data=df)
```

```python
plt.ticklabel_format(style='plain',axis='both')
plt.style.use({'figure.figsize':(32, 8)})
sns.set(font_scale=2)
hue=df['Jim_Density']
sns.scatterplot(x="Urban population", y="Rural population",s=1200,color='g',hue=hue,data=df)
```

```python
plt.style.use({'figure.figsize':(32, 8)})
sns.set(font_scale=2)
fig = plt.figure()
ax = Axes3D(fig)
ax.ticklabel_format(style='plain',axis='both')
ax.set_xlabel('Area', fontsize=40,labelpad=60)
ax.set_ylabel('Urban population',fontsize=40,labelpad=60)
ax.set_zlabel('Rural population',fontsize=40,labelpad=60)
ax.scatter(df['Jim_Area'],df['Urban population'], df['Rural population'],s=1600,c='r')
```

```python
plt.style.use({'figure.figsize':(32, 24)})
plt.ticklabel_format(style='plain',axis='both')
sns.set(font_scale=2)
sns.barplot(y="State / Union Territory", x="Jim_Density",data=df)
```

# 病例探索

```python
df = pd.read_csv('./COVID19/IndividualDetails.csv')
df['diagnosed_date'] = pd.to_datetime(df['diagnosed_date'],format='%d/%m/%Y')
df['diagnosed_date'] = df['diagnosed_date'].apply(lambda x : x.strftime('%Y-%m-%d'))
df.isnull().sum()
```

```python
df1 = pd.DataFrame()
a = df.groupby(['diagnosed_date','detected_state'])['id'].count().reset_index(name="count")
a.tail(3)
```

```python
pt = a.pivot_table(index='diagnosed_date', columns='detected_state',  values=['count'],aggfunc=np.sum)
pt.head(3)
```

```python
pt.fillna(value=0,inplace=True)
pt.sort_values('diagnosed_date',ascending=False,inplace=True)
pt.head(3)
```

```python
stats_list = pt['count'].columns.values.tolist() #list(df['count'])
date_list = pt['count'][stats_list[0]]
print(len(date_list),len(stats_list))
```

```python
plt.style.use({'figure.figsize':(64, 64)})
plt.ticklabel_format(style='plain',axis='both')
cmap = sns.cubehelix_palette(start = 1, rot = 3, gamma=0.8, as_cmap = True)
sns.heatmap(pt, cmap = cmap, linewidths = 0.05,annot=False, fmt="g")
```
