# Prepare train data

```python
df = pd.read_csv('./US/us_states_covid19_daily.csv')
#df = pd.read_csv('./US-812/us_states_covid19_daily.csv')
del df['hash']
del df['commercialScore']
del df['negativeRegularScore']
del df['negativeScore']
del df['positiveScore']
del df['score']
del df['grade']
del df['hospitalizedIncrease']
del df['deathIncrease']
del df['hospitalizedCumulative']
del df['hospitalizedCurrently']
del df['inIcuCurrently']
del df['inIcuCumulative']
del df['onVentilatorCurrently']
del df['onVentilatorCumulative']
del df['recovered']
del df['lastUpdateEt']
del df['dateModified']
del df['checkTimeEt']
del df['dateChecked']
df['date'] = df['date'].astype(str)
df.fillna(value=0,inplace=True)
df.head(5)
```

```python
def GetRawSamples():
    df2 = pd.DataFrame()
    df2['date'] = df['date']
    df2['state'] = df['state']
    df2['positive']=df['positive']
    df2['positiveIncrease']=df['positiveIncrease']
    df2['test'] = df['total']
    df2['testIncrease'] = df['totalTestResultsIncrease']
    df2.sort_values('date',ascending=True,inplace=True)
    df2 = df2[df2['date']>='20200401']
    df2 = df2.reset_index()
    del df2['index']
    return df2

def GetSamples(state = None):
    dr = GetRawSamples()
    if state:
        dr = dr[dr['state']==state]
    else:
        dr = dr.groupby(['date'])['positive','positiveIncrease','test','testIncrease'].agg('sum')
        #dr = dr.to_frame().reset_index()
        dr['state'] = 'ALL'
    dr.sort_values('date',ascending=True,inplace=True)
    dr = dr.reset_index()
    return dr
```

```python
#ADF Test result同时小于1%、5%、10%
#P-value (不变显著性) 接近0。
def check_stationarity(timeSeries):
    dftest = adfuller(timeSeries)
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    del dfoutput['#Lags Used']
    del dfoutput['Number of Observations Used']
    return dfoutput
```

```python
df0 = GetSamples()
data = pd.DataFrame()
raw_data = df0['positiveIncrease']
raw_date = df0['date']
log_data = np.log(raw_data)
raw_data.plot()
```

```python
check_stationarity(raw_data)
```

```python
plot_acf(raw_data).show()
plot_pacf(raw_data).show()
```

```python
data_mean = raw_data.rolling(window=7).mean()
data_mean.dropna(inplace=True)
```

```python
check_stationarity(data_mean)
```

```python
data_diff_1 = data_mean.diff(1)
data_diff_1.dropna(inplace=True)
data_diff_2 = data_diff_1.diff(1)
data_diff_2.dropna(inplace=True)
#data_diff_3 = data_diff_2.diff(1)
#data_diff_3.dropna(inplace=True)
#check_stationarity(data_diff_3)
```

# 模型选择

```python
train_data = raw_data.copy()

rol_mean = log_data.rolling(window=7).mean()
rol_mean.dropna(inplace=True)
ts_diff_1 = rol_mean.diff(1)
ts_diff_1.dropna(inplace=True)

ts_diff_2 = ts_diff_1.diff(1)
ts_diff_2.dropna(inplace=True)

def restor_data(predict_ts):
    #date restore
    diff_shift_ts = ts_diff_1.shift(1)
    diff_recover_1 = predict_ts.add(diff_shift_ts)
    
    rol_shift_ts = rol_mean.shift(1)
    diff_recover = diff_recover_1.add(rol_shift_ts)
    rol_sum = log_data.rolling(window=6).sum()
    rol_recover = diff_recover*7 - rol_sum.shift(1)
    log_recover = np.exp(rol_recover)
    rol_recover.dropna(inplace=True)
    log_recover.dropna(inplace=True)
    #log_recover.fillna(value=0,inplace=True)
    return rol_recover
    return log_recover

def show_result(ts,log_recover,error):
    plt.figure(facecolor='white')
    log_recover.plot(color='blue', label='Predict')
    ts.plot(color='red', label='Original')
    plt.legend(loc='best')
    plt.title('RMSE: %.4f'% error)
    plt.show()
```

# Train

```python
def TrainModel(p,q):
    model = ARMA(ts_diff_2, order=(p, q)) 
    result_arma = model.fit(disp=-1, method='css' )
    predict_ts = result_arma.predict()
    log_recover = restor_data(predict_ts)
    ts = train_data[train_data.size-log_recover.size:,]
    ts = np.log(ts)
    error = np.sqrt(sum((log_recover-ts)**2)/ts.size)
    return result_arma,ts,log_recover,error

min_result_arma = None
min_ts= None
min_log_recover= None
min_error = 65536
min_p = None
min_q = None
ARMA_model = ARIMA(train_data, order=(3, 2, 2)).fit(disp=-1)
for p in range(0,5):
    for q in range(0,5):
        try:
            result_arma,ts,log_recover,error = TrainModel(p,q)
            if min_error > error:
                min_error = error
                min_ts = ts
                min_p = p
                min_q = q
                min_log_recover = log_recover
                print("P=%d,Q=%d,Error=%f"%(p,q,error))
        except:
            pass
            
print("Min paramters ",min_p,min_q)
show_result(min_ts,min_log_recover,min_error)
```

# prediction

```python
real_list = [23871,27512,31055,31958,27287]
predict_ts=ARMA_model.forecast(10)  
for i in range(len(real_list)):
    real = real_list[i]
    predicted = int(predict_ts[0][i])
   
    diff = abs(predicted-real)
    rate = "%.4f"%(diff*100/real)
    print("real:",real, " pred :",predicted, " error :",diff," rate :",rate)
```
