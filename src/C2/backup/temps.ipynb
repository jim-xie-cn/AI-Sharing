```python
g_classifiers = [
    # 逻辑回归
    ('Logistic Regression', LogisticRegression()), 
    # K最近邻
    ('Nearest Neighbors', KNeighborsClassifier(3)),  
    # 线性的支持向量机
    ('Linear SVM', SVC(kernel='linear', C=0.025)),
    # 径向基函数的支持向量机
    ('RBF SVM', SVC(gamma=2, C=1)),
    # 基于拉普拉斯近似的高斯过程
    ('Gaussian Process', GaussianProcessClassifier(1.0 * RBF(1.0))),  
    # 决策树
    ('Decision Tree', DecisionTreeClassifier(max_depth=5)),  
    # 随机森林
    ('Random Forest', RandomForestClassifier(max_depth=5, n_estimators=10, max_features=1)),  
    # 通过迭代弱分类器而产生最终的强分类器的算法
    ('AdaBoost', AdaBoostClassifier()),  
    ('Extra Trees', ExtraTreesClassifier()),
    # 梯度提升树
    ('GradientBoosting', GradientBoostingClassifier()),  
    ('Bagging', BaggingClassifier()),
    # 朴素贝叶斯
    ('Naive Bayes', GaussianNB()),  
    # 二次判别分析
    ('QDA', QuadraticDiscriminantAnalysis()),  
    # 线性判别分析
    ('LDA', LinearDiscriminantAnalysis()),  
    # 多层感知机
    ('MLP', MLPClassifier(alpha=1)),
    # 极端梯度提升
    ('XGB', XGBClassifier()),
]
```

```python
def GetParityData():
    dataset_size = 100
    x1 = np.random.random_integers(1,100,dataset_size )
    x2 = np.random.random_integers(1,100,dataset_size )
    y = []
    x = []
    for t1,t2 in zip(x1,x2):
        if (t1+t2)%2 == 0:
            y.append([1])
            x.append([t1,t2])
        else:
            x.append([t1,t2])
            y.append([0])
    return np.array(x),np.array(y)

def GetParityDataEx():
    dataset_size = 100
    x1 = np.random.random_integers(1,100,dataset_size )
    x2 = np.random.random_integers(1,100,dataset_size )
    y = []
    x = []
    for t1,t2 in zip(x1,x2):
        if (t1+t2)%2 == 0:
            y.append([1])
            x.append([t1,t2])
        else:
            x.append([-t1,-t2])
            y.append([-1])
    return np.array(x),np.array(y)
```

```python
raw_x,raw_y = GetParityData()
test_x,test_y = GetParityData()
```

```python
df = pd.DataFrame(raw_x)
df['label'] = raw_y
plt.style.use({'figure.figsize':(12, 6)})
ax = sns.scatterplot(x=0, y=1, data=df,hue='label',s=300)
ax.ticklabel_format(style='plain',axis='both')
plt.xlabel('x1',fontsize=24)
plt.ylabel('x2',fontsize=24)
```

```python
sess = tf.Session()
result = []
for name, clf in g_classifiers:
    clf.fit(raw_x, raw_y)
    score = clf.score(test_x,test_y)
    result.append((name,score))
    print(name,score)
    continue
    pred_Y = np.array(clf.predict(test_x))
    conf = tf.confusion_matrix(labels=test_y, predictions=pred_Y,num_classes=2)
    conf_numpy = conf.eval(session=sess)
    conf_df = pd.DataFrame(conf_numpy)
df = pd.DataFrame(result)
plt.xlabel('Classifier',fontsize=2)
plt.ylabel('Accuracy',fontsize=2)
sns.barplot(y=0, x=1, data=df)
```
